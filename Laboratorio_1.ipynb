## **Problema 1**

Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la
función debe devolver una imagen la cual tenga activos los canales de color según los siguientes
puntos:
1.  Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color rojo.
2.  Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
3.  Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color azul.
4.  Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y
verde.
5.  Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y
azul.
6.  Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y
rojo.

```python
import cv2 # importar libreriras a usar
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activar_canales_de_color(imagen_cargada, color):
    imagen_bgr = cv2.imread(imagen_cargada) # Leer la imagen con OpenCV 
    imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB) # transformar de bgr a rgb

    
    mask = np.zeros_like(imagen_rgb) # Crear una máscara para activar los canales de color deseados

    if color == 1:  # Mostrar únicamente el color azul
        mask[:, :, 0] = imagen_rgb[:, :, 0]  # Copiar el canal azul
    elif color == 2:  # Mostrar únicamente el color verde
        mask[:, :, 1] = imagen_rgb[:, :, 1]  # Copiar el canal verde
    elif color == 3:  # Mostrar únicamente el color rojo
        mask[:, :, 2] = imagen_rgb[:, :, 2]  # Copiar el canal rojo
    elif color == 10:  # Mostrar únicamente los colores rojo y verde
        mask[:, :, 0] = imagen_rgb[:, :, 0]  # Copiar el canal azul
        mask[:, :, 1] = imagen_rgb[:, :, 1]  # Copiar el canal verde
    elif color == 20:  # Mostrar únicamente los colores verde y azul
        mask[:, :, 1] = imagen_rgb[:, :, 1]  # Copiar el canal verde
        mask[:, :, 2] = imagen_rgb[:, :, 2]  # Copiar el canal rojo
    elif color == 30:  # Mostrar únicamente los colores azul y rojo
        mask[:, :, 0] = imagen_rgb[:, :, 0]  # Copiar el canal azul
        mask[:, :, 2] = imagen_rgb[:, :, 2]  # Copiar el canal rojo
    else:
        # Si el valor de "color" no coincide con ninguna de las opciones anteriores,
        # devolver la imagen original sin cambios
        return imagen_rgb

    # usa máscara a la imagen original para activar los canales de color deseados
    imagen_modificada = cv2.bitwise_and(imagen_rgb, mask)

    return imagen_modificada
```

# Uso de la funcion

```python
#Valores de la funcion

imagen_cargada = 'L1-p1.jpg'
canal_deseado = 1

# Uso de la funcion
imagen_modificada = activar_canales_de_color(imagen_cargada, canal_deseado)

#mostral la imagen
plt.imshow(imagen_modificada)
plt.show()
```

# **Problema 2:**

En el .zip del laboratorio se le compartió un conjunto de imágenes en escala de grises, estas imágenes
fueron creadas utilizando una escala de grises en 3D, cree una función que dadas las 3 imágenes seconstruya la imagen original a color.

```python
#def combinar_imagenes_grises_a_color(imagen_rojo):
def combinar_imagenes_grises_a_color(imagen_azul, imagen_verde, imagen_rojo ):
    # Leer las imágenes en escala de grises
    
    canal_azul = cv2.imread(imagen_azul, cv2.IMREAD_GRAYSCALE)
    canal_verde = cv2.imread(imagen_verde, cv2.IMREAD_GRAYSCALE)
    canal_rojo = cv2.imread(imagen_rojo, cv2.IMREAD_GRAYSCALE)
    
    

    # Obtener las dimensiones de las imágenes
    alto, ancho = canal_azul.shape

    # Crear una matriz para almacenar la imagen a color
    imagen_color = np.zeros((alto, ancho, 3), dtype=np.uint8)

    # Asignar cada canal a la imagen a color
    imagen_color[:, :, 0] = canal_azul  # Canal azul
    imagen_color[:, :, 1] = canal_verde  # Canal verde
    imagen_color[:, :, 2] = canal_rojo  # Canal rojo

    return imagen_color
```

# **Problema 3:**

Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en
cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir
una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la
imagen del Problema #2.

```python
def crear_escala_de_grises_tres_dimensiones(imagen):
    # Leer la imagen a color
    imagen_color = cv2.imread(imagen)

    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)

    # Crear tres imágenes con el mismo tamaño que la imagen original en escala de grises
    imagen_rojo = np.zeros_like(imagen_gris)
    imagen_verde = np.zeros_like(imagen_gris)
    imagen_azul = np.zeros_like(imagen_gris)

    # Copiar los valores de píxeles de la imagen en escala de grises a los canales correspondientes
    imagen_rojo[:, :] = imagen_gris
    imagen_verde[:, :] = imagen_gris
    imagen_azul[:, :] = imagen_gris

    return imagen_rojo, imagen_verde, imagen_azul

# Ruta de la imagen que deseas convertir a escala de grises en tres dimensiones
imagen_ruta = 'L1-p1.jpg'

# Crear la escala de grises en tres dimensiones
imagen_rojo, imagen_verde, imagen_azul = crear_escala_de_grises_tres_dimensiones(imagen_ruta)

# Mostrar las imágenes resultantes en escala de grises utilizando plt.imshow()
plt.imshow(imagen_rojo, cmap='gray')
plt.title('Canal Rojo (Escala de Grises)')
plt.show()

plt.imshow(imagen_verde, cmap='gray')
plt.title('Canal Verde (Escala de Grises)')
plt.show()

plt.imshow(imagen_azul, cmap='gray')
plt.title('Canal Azul (Escala de Grises)')
plt.show()
```

# **Problema 4:**
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media
de la distribución.

# Use la imagen del problema 1 y 3

```python
def histograma_con_media(imagen):
    # Leer la imagen a color
    imagen_color = cv2.imread(imagen)

    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)

    # Obtener los valores de píxeles de cada canal de color
    valores_rojo = imagen_color[:, :, 2].ravel()
    valores_verde = imagen_color[:, :, 1].ravel()
    valores_azul = imagen_color[:, :, 0].ravel()

    # Calcular el promedio aritmético de la escala de grises
    promedio_gris = np.mean(imagen_gris)

    # Crear y mostrar el histograma de cada canal de color
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 2, 1)
    plt.hist(valores_rojo, bins=256, color='red', alpha=0.7)
    plt.axvline(np.mean(valores_rojo), color='black', linestyle='dashed', linewidth=2)
    plt.title('Histograma del Canal Rojo')
    
    plt.subplot(2, 2, 2)
    plt.hist(valores_verde, bins=256, color='green', alpha=0.7)
    plt.axvline(np.mean(valores_verde), color='black', linestyle='dashed', linewidth=2)
    plt.title('Histograma del Canal Verde')
    
    plt.subplot(2, 2, 3)
    plt.hist(valores_azul, bins=256, color='blue', alpha=0.7)
    plt.axvline(np.mean(valores_azul), color='black', linestyle='dashed', linewidth=2)
    plt.title('Histograma del Canal Azul')

    # Crear y mostrar el histograma de escala de grises
    plt.subplot(2, 2, 4)
    plt.hist(imagen_gris.ravel(), bins=256, color='gray', alpha=0.7)
    plt.axvline(promedio_gris, color='black', linestyle='dashed', linewidth=2)
    plt.title('Histograma de Escala de Grises')

    plt.tight_layout()
    plt.show()

# Ruta de la imagen que deseas analizar
imagen_ruta = 'L1-p1.jpg'

# Mostrar los histogramas
histograma_con_media(imagen_ruta)
```

# El modelo de color HSV

El modelo de color HSV se basa en un sistema de color en el que el espacio de color se representa mediante un único cono. Los tres componentes del cono son tono, saturación y valor, como se muestra en la siguiente ilustración:

```python
img = cv2.imread("HSV.png")
plt.imshow(img)
plt.show()
```

En el modelo de color HSV, se cumple lo siguiente:

1. Tono: especifica el tono (color) establecido para el color. El tono se indica como un entero del 0 al 240, ambos incluidos. Este se remuestrea a partir de los valores de 0° a 360° en que se indica el tono como ángulo en sentido antihorario alrededor del cono de color. Los colores primarios y secundarios presentan los siguientes valores de tono: rojo = 0 (0°), amarillo = 40 (60°), verde = 80 (120°), cian = 120 (180°), azul = 160 (240°) y magenta = 201 (300°).

2. Saturación: especifica la intensidad de saturación establecida para el color. La saturación se indica como un entero del 0 al 255, ambos incluidos (que representan del 0 al 100%). La saturación de un color hace referencia a la medida en que se aleja de un color neutro como el gris o, dicho de un modo más sencillo, su colorido. Cuando la saturación es de 255, el color presenta una saturación máxima. Cuando el valor de la saturación es 0, el color es no saturado y parece gris (a menos que el valor sea 0 ó 255, en cuyo caso aparece negro o blanco).

3. Brillo: especifica la intensidad del blanco en el color. El valor se indica como un entero del 0 al 255, ambos incluidos (que representan del 0 al 100%). Un color con un valor establecido en 0 aparece negro. Un color con un valor establecido en 255 y una saturación de 0 aparece como blanco.

La ventaja del espacio de color HSV radica en que separa la información de color del brillo, lo que facilita la manipulación y el ajuste de los colores sin afectar el brillo de la imagen.

Para mapear colores en el espacio de color HSV, se pueden utilizar diferentes métodos dependiendo del software o biblioteca que se utilice. En la biblioteca OpenCV de Python, el mapeo de colores a HSV se realiza utilizando la función cv2.cvtColor() con los respectivos códigos de conversión.
